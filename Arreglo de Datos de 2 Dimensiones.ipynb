```python
#Analistica de Datos LTIN
#Luis Angel Arias Barajas
#Arreglo de Datos de 2 dimensiones

#Importando las bibliotecas necesarias
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

#Generando datos aleatorios
np.random.seed(42) #Para reproducibilidad
data = np.random.randn(100) #Puntos de datos aleatorios de una distribuci√≥n normal

#Creando un Dataframe con pandas
df = pd.DataFrame(data, columns=['valores'])

#Calculando estadistica descriptiva con scipy y pandas
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
std_dev=np.std(data)

#Imprimir estadisticas descriptivas
print(f"Media: {mean}")
print(f"Mediana: {median}")
print(f"Moda {mode.mode[0]} (aparece {mode.count[0]} veces")
print(f"Desviaci√≥n Estandar: {std_dev}")
      
#Visualizando los datos con matplotlib
plt.figure(figsize=(10,6))
plt.hist(data, bins=20, color='blue', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label='media')
plt.axvline(median, color='yellow', linestyle='dotted', linewidth=1, label='Mediana')
plt.axvline(mode.mode[0], color='green', linestyle='dotted', linewidth=1, label='Moda')
plt.title('Distribuci√≥n de los datos')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
#Arreglo de Datos de 3 dimensiones

#Importar las librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Generar datos tridimensionales:
"Vamos a asumir que quieres una matr√≠z 10x10x10. Esto se puede adaptar seg√∫n tus necesidades"

x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
z = np.linspace(0,10,10)


x,y,z = np.meshgrid(x,y,z)
values = np.random.rand(10,10,10) #Este es el arreglo de datos aleatorios

#Crear un Dataframe de pandas con los datos:
df = pd.DataFrame({
'x': x.flatten(),
'y': y.flatten(),
'z': z.flatten(),
'values': values.flatten()
})

#Visualizar los datos en 3D usando matplotlib
fig = plt.figure(figsize=(10,7))
ax=fig.add_subplot(111,projection='3d')

sc = ax.scatter(df['x'], df['y'], df['z'], c=df['values'], cmap='viridis', s=50)
plt.colorbar(sc)

ax.set_xlabel('x Label')
ax.set_ylabel('y Label')
ax.set_zlabel('z Label')

ax.set_title('3D Data Visualization with matplotlib')

plt.show()
```
